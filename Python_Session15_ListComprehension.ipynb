**Method-2**
- for loop with if condition

```python
# [<output> <for loop>]
# [<output> <for loop> <if condition>]
```

```python
l1=['hyd','chennai','mum#ai','p#ne']
l2=[i for i in l1 if '#' in i]
l2
```

```python
l1=['hyd','chennai','mum#ai','p#ne']
l2=[i for i in l1 if '#' not in i]
l2
```

```python
l1=[100,20,27,38,47,52,87]
even=[i for i in l1 if i%2==0]
even
```

```python
l1=[100,20,27,38,47,52,87]
odd=[i for i in l1 if i%2!=0]
odd
```

**Method-3: for loop -if-else**

```python
#method-3: [<if output> <if condition> else <else output> <for loop>]
```

```python
l1=[100,20,27,38,47,52,87]
l2=[f'even :{i}' if i%2==0 else f'odd :{i}' for i in l1]
l2
```

```python
l=[45,76,89,53,67,90,82,49,33,99]
#filter the students who passed marks >= 50
#calculate grade of all stud 
# A: 85+, B: 70-84 ,C: 50-69 D:<50
#Add 5 grace to stud who failed <50
#find the average marks of the class
```

```python
l = [45, 76, 89, 53, 67, 90, 82, 49, 33, 99]
passed_stud = [i for i in l if i >= 50]
grade = []
for i in l:
    if i >= 85:
        grade.append('A')
    elif i >= 70:
        grade.append('B')
    elif i >= 50:
        grade.append('C')
    else:
        grade.append('D')
update = [i + 5 if i < 50 else i for i in l]
total_marks=sum(l)
average = total_marks / len(l)
print("Passed Students:", passed_stud)
print("Grades:", grade)
print("Updated Marks with Grace:", update)
print("Average Marks:", average)
```

**walrus operator**
- in list comprehension = will not work
- if you want to write equal to (=), then will use walrus operator:=

```python
summ=0
[summ:=summ+i for i in l]
```

```python
total=0
ans=[total:=total+i for i in l]
total/len(ans)
```
