$pattern-1$
- Function with one arguement

```python
#Syntax

# function name = Lambda <arguement name>:<return output>
#in above 
#function name : mul
#arguement name : a
#return output : a*a
```

```python
mul=lambda a : a*a
mul(10)
```

```python
def cube(x):
    return(x**3)
cube(10)
```

```python
cube=lambda x:x**3
cube(10)
```

$Pattern-2$
- function with two arguements

```python
# syntax
# <function name>= lambda <arg1>,<arg2>: return(output)
```

```python
def add(a,b):
    return(a+b)
add(100,200)
```

```python
add1=lambda a,b:a+b
add1(200,800)
```

```python
def avg(a,b,c):
    return(round((a+b+c)/3,2))
avg(99.99,99,99)
```

```python
avg1=lambda a,b,c:round((a+b+c)/3,3)
avg1(99.90,99,99)
```

$Pattern-3$
- default arguement

```python
avg2=lambda a,b,c=500:round((a+b+c)/3,3)
avg2(99.90,99,99)
```

$Pattern-4$
- if else

```python
def max(a,b):
    if a>b:
        return(a)
    else:
        return(b)
max(10,20)
```

```python
#Syntax
#Function name= lambda <arg1>,<arg2>: <if out> <if con> else <elseout>
```

```python
max=lambda a,b:a if a>b else b
max(99,100)
```

**map**
- map is used to combine the function and iterator

```python
#Step-3:
#map(lambda <variable>: <op>,,iterator>)
list(map(lambda i:i.capitalize(),list1))
```

```python
tuple(map(lambda i:i.capitalize(),list1))
```

```python
l1=['Hyd', 'Mumbai', 'Pune']
list(map(lambda i:i.upper(),l1))
```

```python
li1=['hyd','mumbai','p#ne','de#hi']
list(map(lambda i: '#' in i,li1))
```

**filter**
- whenever condition statements
- which means we are filtering the answers based on condition
- map will give boolean outputs, True or False
- True answers we can see by applying filter only

```python
li1=['hyd','mumbai','p#ne','de#hi']
list(filter(lambda i: '#' in i,li1))
```

```python
l2='hello how are you'
ans=list(map(lambda i:i.title(),l2.split()))
ans
```

**reduce**
- reduce is a method to write all inbuilt functions using lambda method
- it is available in a package called **func tools**
- level-1: reduce(lambda sum,i: sum+i,list1)
- level-2: reduce(lambda sum,i: sum+i,list1,value)
- in level-1 by default sum=0
- in level-2 the sum starts with some value

```python
import functools 
list1=[1,2,3,4,5]
functools.reduce(lambda sum,i: sum+i,list1,200)
```

```python
import functools 
list1=[1,2,3,4,5]
functools.reduce(lambda val,i: val*i,list1)
```

```python
from functools import reduce
list1=[1,2,3,4,5]
reduce(lambda val,i: val*i,list1)
```

```python
import functools as ft
list1=[1,2,3,4,5]
ft.reduce(lambda val,i: val*i,list1)
```
