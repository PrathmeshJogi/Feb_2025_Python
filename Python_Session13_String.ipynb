```python
str1='virat.kohli@rcb.com'
str2='rohit.sharma@mi.com'
str3='m.dhoni@chennai.com'
str4='omkar.nallagoni@cts.com'
str5='nandu.kishore@google.com'
#get firstname ,secondname, cname

#idea
#find the first dot index
#find @ index
#find the second dot index

def extrac(str1):
    first_dot = str1.find('.')
    at_symbol = str1.find('@')
    firstname1 = str1[:first_dot]
    lastname1 = str1[first_dot + 1:at_symbol]
    company1 = str1[at_symbol + 1:str1.find('.', at_symbol)]
    print(firstname1,lastname1,company1)
    
extrac('virat.kohli@rcb.com')
```
